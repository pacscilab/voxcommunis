```python
import pandas as pd
import numpy as np
```

```python
lang_ver19 = pd.read_csv('VoxCommunis_Info.csv', usecols=['code_cv', 'name_cv', 'totalHrs', 'validHrs'])
lang_ver19.rename(columns = {'clips': 'clips_v19', 
                             'validClips': 'validClips_v19',
                             'totalHrs': 'totalHrs_v19',
                             'validHrs': 'validHrs_v19'},
                             inplace=True)
lang_ver19.drop_duplicates(inplace=True)
lang_ver19
```

```python
lang_ver16 = pd.read_csv('/Users/miaozhang/switchdrive/vxc_scripts/vxc_pipeline/VoxCommunis_Info_v16.csv', usecols=['code_cv', 'name_cv', 'totalHrs', 'validHrs'])
lang_ver16.rename(columns = {'clips': 'clips_v16', 
                             'validClips': 'validClips_v16',
                             'totalHrs': 'totalHrs_v16',
                             'validHrs': 'validHrs_v16'},
                             inplace=True)
lang_ver16.drop_duplicates(inplace=True)
lang_ver16
```

```python
merged = pd.merge(lang_ver16, lang_ver19, how = 'outer', on = ['code_cv', 'name_cv'])
merged['diff_validhrs_19_16'] = round(merged['validHrs_v19'].fillna(0) - merged['validHrs_v16'].fillna(0), 2)
merged.sort_values(by = ['diff_validhrs_19_16'], ascending = False)
print(merged)
merged.to_csv('VoxCommonis_ver_comp.csv')
```
